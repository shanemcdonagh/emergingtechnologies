<h2 style="color: rgb(0, 91, 94);">Big O Notation</h2>

<hr style="border-top: 1px solid rgb(0, 91, 94);" />

Let $f$ and $g$ be functions. Then $f$ is $O(g)$ if there are positive integers $c$ and $n_0$ such that for every $n \geq n_0$: $f(n) \leq c g(n)$

<h4 style="color: rgb(0, 91, 94)">Explaining Big O Notation</h4>
<hr style="border-top: 1px solid rgb(0, 91, 94);" />

To be able explain the notations and functions, first we must have a general understanding of what <b>Big O Notation</b> is and what it represents. Big O notation is considered to be a metric in computing which allows for us to determine how efficient an algorithm is while the input size continues to grow. <sup><a href="#references">[1]</a></sup> 

In this instance, Big-O notation is used to determine the <u>worst case scenario</u>, time-complexity wise. This can be described as "upper-bound".

Since the running time of an algorithm can vary, it is convenient to have a notation which allows us to specify that the most it can grow is by a certain amount, however it can grow more gradually. This is why Big-O is used.


<h4 style="color: rgb(0, 91, 94);">Explaining the functions</h4>
<hr style="border-top: 1px solid rgb(0, 91, 94);"/>

<p><b>$f(n) = O(g(n))$</b>: The function ùëì = ùëÇ(ùëî) (that Big O is upper bound) is true if there are two positive constants ($c$ and $n_0$) such that....</p>

<p>$n \geq n_0$: </p> Used to express the inequality where this notation becomes true
<ul>
    <i>Example: $3n + 8n + 4 \leq 3n + 8n + 4n$</i>
    <p>Where n = 0, the statement above is <u>false</u> ($f(0) > cg(0)$ in this instance)</p>
    <p>Therefore in this instance, $n_0$ would be 1, meaning any number above 1 gives us our inequality. ($f(1) \leq cg(1)$)
</ul>

<p><b>$f(n) \leq c g(n)$</b>: The function $f$ must be asymptotically less than or
equal to $g$ <sup><a href="#references">[2]</a></sup></p> 
    <ul>
         <i>Example:</i> $3n + 3 \leq 11n$
    </ul>
 


<p>We can prove that $f$ is $O(g)$ by the following:</p>
    <ol>
    <li>Write function $f$ and $g$ where $f \leq g$</li>
    <li>Plot the functions where $n$ increases incrementally</li>
    <li>Review results</li>
    </ol>
<p>If function $g$ has the same or worse time complexity as $f$ ($n$,$nlogn$,$n^2$....), then this proves the statment that $f(n) = O(g(n))$

Since Big O is upper-bound, any time complexity worse than or equal to $O(n)$ proves the statement that $f$ is $O(g)$
</p>


```python
# Video reference 1: https://www.youtube.com/watch?v=A03oI0znAoc
# Video reference 2: https://www.youtube.com/watch?v=dNDMjSwZRE4


# Example 1: 
# f = 2n + 3 ‚â§ 10n (10 being c and n being ùëî(ùëõ))
# To prove: 2n + 3 = O(n)

# Create a figure.
fig, ax = plt.subplots(figsize=(12, 6))

# n values.
n = np.linspace(0.0, 3.0, 1000)

# Plot f(n).
ax.plot(n, (2*n)+3, label="$2n + 3$")

# Plot cg(n).
ax.plot(n, 10*n, label="$10n$")

# Legend.
ax.legend()

# Show.
plt.show();
```

Although $10n$ initally had a slower growth, it eventually started to grow exponentially faster, meaning that $2n + 3 \leq 10n$.

```python
# Example 2: 
# f = 3n^2 + 50 ‚â§ 5n^2 (5 being c and n^2 being ùëî(ùëõ))
# To prove: 3n^2 + 50 = O(n^2)

# Create a figure.
fig, ax = plt.subplots(figsize=(12, 6))

# n values.
n = np.linspace(0.0, 10.0, 100)

# Plot f(n).
ax.plot(n, (3*n)**2+50, label="$3n^2 + 50$")

# Plot cg(n).
ax.plot(n, (5*n)**2, label="$5n^2$")

# Legend.
ax.legend()

# Show.
plt.show();
```

Initially the growth rate of both functions were extremely similiar, however over time $5n^2$ rapidly shot upwards, whereas $3n^2 + 50$ gradually become to smoothen out. This proves that $3n^2 + 50 \leq 5n^2$.

```python
# Example 3: 
# f = 3n^5 + 5 ‚â§ 7n^5 (7 being c and n^5 being ùëî(ùëõ))
# To prove: 3n^5 + 5 = O(n^5)

# Create a figure.
fig, ax = plt.subplots(figsize=(12, 6))

# n values.
n = np.linspace(0.0, 2.0, 100)

# Plot f(n).
ax.plot(n, (3*n)**5+5, label="$3n^5 + 5$")

# Plot cg(n).
ax.plot(n, (7*n)**5, label="$7n^5$")

# Legend.
ax.legend()

# Show.
plt.show();
```

Within this plot we notice that the initial growth of both functions keep at a relatively slow growth rate since the running time is still extremly low. However, we can see the growth in our function $g$ ($10n^5$) spike dramatically, proving that $3n^5 + 5 \leq 7n^5$.

```python
# Example 4: 
# f = 1n + 5 ‚â§ 2n (2 being c and n being ùëî(ùëõ))
# To prove: 1n = O(n)

# Create a figure.
fig, ax = plt.subplots(figsize=(12, 6))

# n values.
n = np.linspace(0.0, 6.0, 1000)

# Plot f(n).
ax.plot(n, (1*n) + 5, label="$1n + 5$")

# Plot cg(n).
ax.plot(n, (2*n), label="$2n$")

# Legend.
ax.legend()

# Show.
plt.show();
```

For the two functions $f$ and $g$, I used lower values since all the other examples were dealing with higher values, so this plot deals with a lower input size. The function $g$ continously increases at a faster rate than $f$, resulting in $1n + 5 \leq 2n$.

```python
# Example 5: 
# f = 10n^2 + 10n ‚â§ 25n^2 (25 being c and n being ùëî(ùëõ))
# To prove: 10n^2 + 10n = O(n^2)

# Create a figure.
fig, ax = plt.subplots(figsize=(12, 6))

# n values.
n = np.linspace(0.0, 5.0, 100)

# Plot f(n).
ax.plot(n, (10*n)**2 + (10*n), label="$10n^2 + 10n$")

# Plot cg(n).
ax.plot(n, (25*n)**2, label="$25n^2$")

# Legend.
ax.legend()

# Show.
plt.show();
```

Within this example, really early on it is quite evident that the growth of function $g$ is growing much more vapidly than function $f$, therefore proving the statement that $10n^2 + 10n \leq 25n^2$.

<h4 style="color: rgb(0, 91, 94);">Exercise: Conclusion</h4>
<hr style="border-top: 1px solid rgb(0, 91, 94);" />

Since we were able to prove with the above plots that $f(n) \leq g(n)$ (that $f(n)$ is asymptotically less than or equal to g(n)), this provides evidence that $f$ is $O(g)$.

<h4 style="color: rgb(0, 91, 94);">Exercise 2: Turing Machine Explanation</h4>
<hr style="border-top: 1px solid rgb(0, 91, 94);" />

To fully understand the code displayed below, it is first important to be able to know what a Turing Machine is, within the context of theoretical computer science.

A Turing Machine is described to be a "hypothetical" machine invented by Alan Turing in the year 1936. Although it can be considered simplistic in nature, a TM can simulate any type of computer algorithm, no matter how complex.<sup><a href="#references">[3]</a></sup></p> 

The Turing Machine consists of the following: 

<ul>
    <li>An infinitely-long <b><i>tape</i></b> (acts as memory in a standard computer) consisting in this scenario, 3 symbols (0,1, and " ")</li> 
    <li>A <b><i> read-write head</i></b> to read the symbols on the tape</li>
    <li>A <b><i>finite state table</i></b> to inform the machine what state to be transition to based on current symbol and state</li>
</ul>

The mathematical definition of a Turing Machine can be describe as a quadruple $T=(Q,Œ£,s,Œ¥)$ where:
<ul>
    <li>$Q$ is a finite set of states $q$ that the machine can transition between</li>
    <li>$Œ£$ is a finite set of symbols in which the machine can process</li>
    <li>$s$ is the initial state $s‚ààQ$</li>
    <li>Œ¥ is a transition function determining the next move <sup><a href="#references">[4]</a></sup></li>
</ul>

This definition can be broken down into the following steps:
<ul>
    <li>The read-write head reads our tape 1 symbol at a time (1 symbol per square)</li>
    <li>The read-write head next action is determined by the state table, based on the current state and symbol</li>
    <li>The head performs its action and then transitions to a new state</li>
    <li>This continues until the machine reaches a halting state</li>
</ul>

Below, is an example of a state table, displayed as a <b>diagram</b>:

<img width="500px" src="https://www.cl.cam.ac.uk/projects/raspberrypi/tutorials/turing-machine/fsm0.jpg">
<center><a href="https://www.cl.cam.ac.uk/projects/raspberrypi/tutorials/turing-machine/fsm0.jpg">Image Source</a></center>

<br>
The following state diagram has two states ($S1$ and $S2$) and 3 symbols (a "blank" state, 0, and 2).

The transitions are as follows:
<ul>
    <li><i>If</i> state is 0 <i>and</i> the symbol is "Blank"  =>  Move the tape to the left</li>
    <li><i>If</i> state is 0 <i>and</i> the symbol is '0'  =>  Write '1' and move tape to the right</li>
    <li><i>If</i> state is 0 <i>and</i> the symbol is "1"  =>  Write '0' and move tape to the right</li>
    <br>
    <li><i>If</i> state is 1 <i>and</i> the symbol is "Blank"  =>  Write 'Blank', move tape to the right and halt operation</li>
    <li><i>If</i> state is 1 <i>and</i> the symbol is '0'  =>  Write '1' and move tape to the left</li>
    <li><i>If</i> state is 1 <i>and</i> the symbol is '1'  =>  Write '0' and move tape to the left</li>
</ul>

<u>To display this with code, we shall create a simple Turing Machine consisting of 3 symbols below:</u>

```python
# State table.
table = {
    ('X', '_'): ['_', 'R', 'T'],
    ('X', '0'): ['0', 'R', 'X'],
    ('X', '1'): ['1', 'R', 'Y'],
    ('Y', '_'): ['_', 'R', 'F'],
    ('Y', '0'): ['0', 'R', 'Y'],
    ('Y', '1'): ['1', 'R', 'X'],
}

# Tape input.
tape = list('01110001')
# Position on tape.
pos = 0
# Initial state is first in table.
state = 'X'

#if tape.count('1') > 0:
    #state = 'T';
    #print("Tape input contains 1")

# Keep going while we are not in a halting state.
while state not in ['T', 'F']:
    # Print the current status.
    print(''.join(tape[:pos]) + state + ''.join(tape[pos:]))
    # Get the row of the table.

    row = table[(state, tape[pos])]
    
    #if row[0] == '1':
        #tape = tape + ['_']
    #else:
    # Overwrite the symbol.
    tape[pos] = row[0]
    # Move left or right.
    if row[1] == 'R':
        # Put blanks on tape as necessary.
        if pos == len(tape) - 1:
            tape = tape + ['_']
        # Increase position.
        pos = pos + 1
    else:
        # Put blanks on tape as necessary.
        if pos == 0:
            tape = ['_'] + tape
            # The position on the tape has to move with it.
            pos = pos + 1
        # Increase position.
        pos = pos - 1
    # Update the state.
    state = row[2]

# Print the current status once we are in a halting state.
print(''.join(tape[:pos]) + state + ''.join(tape[pos:]))
```

<h4 style="color: rgb(0, 91, 94)">Problem</h4>

Change the above code to only accept inputs that do not contain any 1's

<h4 style="color: rgb(0, 91, 94)">Solution</h4>

The solution to the above exercise is quite straightforward. After our explanation of a Turing Machine, and specifically what the the presence of a state table establishes, we can simply change our state table to instead accept a different input than 1.

As seen below, we have a state table which accepts 3 inputs (<i>3-symbol Turing Machine</i>):
<ul>
    <li>_ (represents a blank)</li>
    <li>0</li>
    <li>2</li>
</ul>

By changing the first state table from the original example, we can specify for the state table to only use these specific inputs and the finite number of states the machine can transition to based on its current state and symbol.

By this logic, we can add more transitions and states for the machine to be in. For example, let's try to add a $Z$ state to the table, which will write a new symbol (or erase it) and continue to a new state.

```python
# State table.
table = {
    ('X', '_'): ['_', 'R', 'T'],
    ('X', '0'): ['0', 'R', 'Z'],
    ('X', '2'): ['2', 'R', 'Y'],
    ('Y', '_'): ['_', 'R', 'F'],
    ('Y', '0'): ['0', 'R', 'Z'],
    ('Y', '2'): ['2', 'R', 'X'],
    ('Z', '_'): ['_', 'R', 'G'],
    ('Z', '0'): ['0', 'R', 'X'],
    ('Z', '2'): ['2', 'R', 'Y'],
}

# Tape input.
tape = list('02220020')
# Position on tape.
pos = 0
# Initial state is first in table.
state = 'X'

#if tape.count('1') > 0:
    #state = 'T';

# Keep going while we are not in a halting state.
while state not in ['T', 'F', 'G']:
    # Print the current status.
    print(''.join(tape[:pos]) + state + ''.join(tape[pos:]))
    # Get the row of the table.

    row = table[(state, tape[pos])]
    
    #if row[0] == '2':
        #tape = tape + ['_']
    #else:
    # Overwrite the symbol.
    tape[pos] = row[0]
    # Move left or right.
    if row[1] == 'R':
        # Put blanks on tape as necessary.
        if pos == len(tape) - 1:
            tape = tape + ['_']
        # Increase position.
        pos = pos + 1 
    else:
        # Put blanks on tape as necessary.
        if pos == 0:
            tape = ['_'] + tape
            # The position on the tape has to move with it.
            pos = pos + 1
        # Increase position.
        pos = pos - 1
    # Update the state.
    state = row[2]

# Print the current status once we are in a halting state.
print(''.join(tape[:pos]) + state + ''.join(tape[pos:]))
```

<h4 style="color: rgb(0, 91, 94);">Exercise: Conclusion</h4>
<hr style="border-top: 1px solid rgb(0, 91, 94);" />

As seen above we added a new $Z$ state to the table, adding to the finite set of states $Q$. We also added more Œ¥ transition functions which determine the next state to transition to.

The above output ensures to stop if a blank space occurs no matter what state the machine is initially in ($X,Y,Z$), by setting them into halting states ($T,F,G$).

Since updating our state table, it no longer accepts inputs of 1.
