<h2 style="color: rgb(0, 91, 94);">Big O Notation</h2>

<hr style="border-top: 1px solid rgb(0, 91, 94);" />

Let $f$ and $g$ be functions. Then $f$ is $O(g)$ if there are positive integers $c$ and $n_0$ such that for every $n \geq n_0$: $f(n) \leq c g(n)$

<h4 style="color: #001a79;">Exercise 1: Explaining Big O Notation</h4>
<hr style="border-top: 1px solid #001a79;" />

To be able explain the notations and functions, first we must have a general understanding of what <b>Big O Notation</b> is and what it represents. Big O notation is considered to be a metric in computing which allows for us to determine how efficient an algorithm is while the input size continues to grow. https://flexiple.com/algorithms/big-o-notation-cheat-sheet/

In this instance, Big-O notation is used to determine the <u>worst case scenario</u>, time-complexity wise. This can be described as "upper-bound".

Since the running time of an algorithm can vary, it is convenient to have a notation which allows us to specify that the most it can grow is by a certain amount, however it can grow more gradually. This is why Big-O is used.


<h4 style="color: #001a79;">Exercise 1: Explaining the functions</h4>
<hr style="border-top: 1px solid #001a79;" />

<p><b>$f(n) = O(g(n))$</b>: The function ùëì = ùëÇ(ùëî) (that Big O is upper bound) is true if there are two positive constants ($c$ and $n_0$) such that....</p>

<p>$n \geq n_0$: </p> Used to express the inequality where this notation becomes true
<ul>
    <i>Example: $3n + 8n + 4 \leq 3n + 8n + 4n$</i>
    <p>Where n = 0, the statement above is <u>false</u> ($f(0) > cg(0)$ in this instance)</p>
    <p>Therefore in this instance, $n_0$ would be 1, meaning any number above 1 gives us our inequality. ($f(1) \leq cg(1)$)
</ul>

<p><b>$f(n) \leq c g(n)$</b>: The function $f$ must be asymptotically less than or
equal to $g$</p> http://www.columbia.edu/~cs2035/courses/ieor6614.S11/algal.pdf
    <ul>
         <i>Example:</i> $3n + 3 \leq 11n$
    </ul>
 


<p>We can prove that $f$ is $O(g)$ by the following:</p>
    <ol>
    <li>Write function $f$ and $g$ where $f \leq g$</li>
    <li>Plot the functions where $n$ increases incrementally</li>
    <li>Review results</li>
    </ol>
<p>If function $g$ has the same or worse time complexity as $f$ ($n$,$nlogn$,$n^2$....), then this proves the statment that $f(n) = O(g(n))$

Since Big O is upper-bound, any time complexity worse than or equal to $O(n)$ proves the statement that $f$ is $O(g)$
</p>


```python
# Video reference: https://www.youtube.com/watch?v=A03oI0znAoc
# Video reference: https://www.youtube.com/watch?v=dNDMjSwZRE4
# Note reference: http://www.columbia.edu/~cs2035/courses/csor4231.F11/mathbasics.pdf

# Example 1: 
# f = 2n + 3 ‚â§ 10n (10 being c and n being ùëî(ùëõ))
# To prove: 2n + 3 = O(n)

# Create a figure.
fig, ax = plt.subplots(figsize=(12, 6))

# n values.
n = np.linspace(0.0, 3.0, 1000)

# Plot f(n).
ax.plot(n, (2*n)+3, label="$2n + 3$")

# Plot cg(n).
ax.plot(n, 10*n, label="$10n$")

# Legend.
ax.legend()

# Show.
plt.show();
```

Although $10n$ initally had a slower growth, it eventually started to grow exponentially faster, meaning that $2n + 3 \leq 10n$.

```python
# Example 2: 
# f = 3n^2 + 50 ‚â§ 5n^2 (5 being c and n^2 being ùëî(ùëõ))
# To prove: 3n^2 + 50 = O(n^2)

# Create a figure.
fig, ax = plt.subplots(figsize=(12, 6))

# n values.
n = np.linspace(0.0, 10.0, 100)

# Plot f(n).
ax.plot(n, (3*n)**2+50, label="$3n^2 + 50$")

# Plot cg(n).
ax.plot(n, (5*n)**2, label="$5n^2$")

# Legend.
ax.legend()

# Show.
plt.show();
```

Initially the growth rate of both functions were extremely similiar, however over time $5n^2$ rapidly shot upwards, whereas $3n^2 + 50$ gradually become to smoothen out. This proves that $3n^2 + 50 \leq 5n^2$.

```python
# Example 3: 
# f = 3n^5 + 5 ‚â§ 7n^5 (7 being c and n^5 being ùëî(ùëõ))
# To prove: 3n^5 + 5 = O(n^5)

# Create a figure.
fig, ax = plt.subplots(figsize=(12, 6))

# n values.
n = np.linspace(0.0, 2.0, 100)

# Plot f(n).
ax.plot(n, (3*n)**5+5, label="$3n^5 + 5$")

# Plot cg(n).
ax.plot(n, (7*n)**5, label="$7n^5$")

# Legend.
ax.legend()

# Show.
plt.show();
```

Within this plot we notice that the initial growth of both functions keep at a relatively slow growth rate since the running time is still extremly low. However, we can see the growth in our function $g$ ($10n^5$) spike dramatically, proving that $3n^5 + 5 \leq 7n^5$.

```python
# Example 4: 
# f = 1n + 5 ‚â§ 2n (2 being c and n being ùëî(ùëõ))
# To prove: 1n = O(n)

# Create a figure.
fig, ax = plt.subplots(figsize=(12, 6))

# n values.
n = np.linspace(0.0, 6.0, 1000)

# Plot f(n).
ax.plot(n, (1*n) + 5, label="$1n + 5$")

# Plot cg(n).
ax.plot(n, (2*n), label="$2n$")

# Legend.
ax.legend()

# Show.
plt.show();
```

For the two functions $f$ and $g$, I used lower values since all the other examples were dealing with higher values, so this plot deals with a lower input size. The function $g$ continously increases at a faster rate than $f$, resulting in $1n + 5 \leq 2n$.

```python
# Example 5: 
# f = 10n^2 + 10n ‚â§ 25n^2 (25 being c and n being ùëî(ùëõ))
# To prove: 10n^2 + 10n = O(n^2)

# Create a figure.
fig, ax = plt.subplots(figsize=(12, 6))

# n values.
n = np.linspace(0.0, 5.0, 100)

# Plot f(n).
ax.plot(n, (10*n)**2 + (10*n), label="$10n^2 + 10n$")

# Plot cg(n).
ax.plot(n, (25*n)**2, label="$25n^2$")

# Legend.
ax.legend()

# Show.
plt.show();
```

Within this example, really early on it is quite evident that the growth of function $g$ is growing much more vapidly than function $f$, therefore proving the statement that $10n^2 + 10n \leq 25n^2$.

<h2 style="color: rgb(0, 91, 94);">Turing Machine</h2>

<hr style="border-top: 1px solid rgb(0, 91, 94);" />

```python
# State table.
table = {
    ('X', '_'): ['_', 'R', 'T'],
    ('X', '0'): ['0', 'R', 'X'],
    ('X', '1'): ['1', 'R', 'Y'],
    ('Y', '_'): ['_', 'R', 'F'],
    ('Y', '0'): ['0', 'R', 'Y'],
    ('Y', '1'): ['1', 'R', 'X'],
}

# Tape input.
tape = list('01110001')
# Position on tape.
pos = 0
# Initial state is first in table.
state = 'X'

# Keep going while we are not in a halting state.
while state not in ['T', 'F']:
    # Print the current status.
    print(''.join(tape[:pos]) + state + ''.join(tape[pos:]))
    # Get the row of the table.
    row = table[(state, tape[pos])]
    # Overwrite the symbol.
    tape[pos] = row[0]
    # Move left or right.
    if row[1] == 'R':
        # Put blanks on tape as necessary.
        if pos == len(tape) - 1:
            tape = tape + ['_']
        # Increase position.
        pos = pos + 1
    else:
        # Put blanks on tape as necessary.
        if pos == 0:
            tape = ['_'] + tape
            # The position on the tape has to move with it.
            pos = pos + 1
        # Increase position.
        pos = pos - 1
    # Update the state.
    state = row[2]

# Print the current status.
print(''.join(tape[:pos]) + state + ''.join(tape[pos:]))
```
