<h2 style="color: rgb(0, 91, 94);">Big O Notation</h2>

<hr style="border-top: 1px solid rgb(0, 91, 94);" />

Let $f$ and $g$ be functions. Then $f$ is $O(g)$ if there are positive integers $c$ and $n_0$ such that for every $n \geq n_0$: $f(n) \leq c g(n)$

<h4 style="color: #001a79;">Exercise 1: Explaining the functions</h4>
<hr style="border-top: 1px solid #001a79;" />

<p><b>$f(n) = O(g(n))$</b>: The function 𝑓 = 𝑂(𝑔) (that Big O is upper bound) is true if there are two positive constants ($c$ and $n_0$) such that....</p>

<p>$n \geq n_0$: </p> Used to express the inequality this notation becomes true
<ul>
    <i>Example: $3n + 8n + 4 \leq 3n + 8n + 4n$</i>
    <p>Where n = 0, the statement above is <u>false</u> ($f(0) > cg(0)$ in this instance)</p>
    <p>Therefore in this instance, $n_0$ would be 1, the number in which gives us our inequality. ($f(1) \leq cg(1)$)
</ul>

<p><b>$f(n) \leq c g(n)$</b>: The function $f$ must be either less than or greater than $g$</p>
    <ul>
         <i>Example:</i> $3n + 3 \leq 11n$
    </ul>
 


<p>We can prove that $f$ is $O(g)$ by the following:</p>
    <ol>
    <li>Write function $f$ and $g$ where $f \leq g$</li>
    <li>Plot the functions where $n$ increases incrementally</li>
    <li>Review results</li>
    </ol>
<p>If function $g$ has the same or worse time complexity as $f$ ($n$,$nlogn$,$n^2$....), then this proves the statment that $f(n) = O(g(n))$</p>


```python
# Video reference: https://www.youtube.com/watch?v=A03oI0znAoc
# Video reference: https://www.youtube.com/watch?v=dNDMjSwZRE4
# Note reference: http://www.columbia.edu/~cs2035/courses/csor4231.F11/mathbasics.pdf

# Example 1: 
# f = 2n + 3 ≤ 10n (10 being c and n being 𝑔(𝑛))
# To prove: 2n + 3 = O(n)

# Create a figure.
fig, ax = plt.subplots(figsize=(12, 6))

# n values.
n = np.linspace(0.0, 3.0, 1000)

# Plot f(n).
ax.plot(n, (2*n)+3, label="$2n + 3$")

# Plot cg(n).
ax.plot(n, 10*n, label="$10n$")

# Legend.
ax.legend()

# Show.
plt.show();
```

```python
# Example 2: 
# f = 3n^2 + 50 ≤ 5n^2 (5 being c and n being 𝑔(𝑛))
# To prove: 3n + 4 = O(n^2)

# Create a figure.
fig, ax = plt.subplots(figsize=(12, 6))

# n values.
n = np.linspace(0.0, 10.0, 100)

# Plot f(n).
ax.plot(n, (3*n)**2+50, label="$3n^2 + 50$")

# Plot cg(n).
ax.plot(n, (5*n)**2, label="$5n^2$")

# Legend.
ax.legend()

# Show.
plt.show();
```

<h2 style="color: rgb(0, 91, 94);">Turing Machine</h2>

<hr style="border-top: 1px solid rgb(0, 91, 94);" />

```python
# State table.
table = {
    ('X', '_'): ['_', 'R', 'T'],
    ('X', '0'): ['0', 'R', 'X'],
    ('X', '1'): ['1', 'R', 'Y'],
    ('Y', '_'): ['_', 'R', 'F'],
    ('Y', '0'): ['0', 'R', 'Y'],
    ('Y', '1'): ['1', 'R', 'X'],
}

# Tape input.
tape = list('01110001')
# Position on tape.
pos = 0
# Initial state is first in table.
state = 'X'

# Keep going while we are not in a halting state.
while state not in ['T', 'F']:
    # Print the current status.
    print(''.join(tape[:pos]) + state + ''.join(tape[pos:]))
    # Get the row of the table.
    row = table[(state, tape[pos])]
    # Overwrite the symbol.
    tape[pos] = row[0]
    # Move left or right.
    if row[1] == 'R':
        # Put blanks on tape as necessary.
        if pos == len(tape) - 1:
            tape = tape + ['_']
        # Increase position.
        pos = pos + 1
    else:
        # Put blanks on tape as necessary.
        if pos == 0:
            tape = ['_'] + tape
            # The position on the tape has to move with it.
            pos = pos + 1
        # Increase position.
        pos = pos - 1
    # Update the state.
    state = row[2]

# Print the current status.
print(''.join(tape[:pos]) + state + ''.join(tape[pos:]))
```
