<h2 style="color: rgb(0, 91, 94);">Big O Notation</h2>

<hr style="border-top: 1px solid rgb(0, 91, 94);" />

Let $f$ and $g$ be functions. Then $f$ is $O(g)$ if there are positive integers $c$ and $n_0$ such that for every $n \geq n_0$: $f(n) \leq c g(n)$

```python
# Five examples of pairs of functions f and g such that f is O(g) 
    # f is O(g) if there are positive integers  and 0 such that for every 0: ()()

# ()(): This essentially means that function f should be always less than or equal to its counterpart
    # Example: If n = 2, then 2  20
# 0: For each term on the left-hand side (n number of terms), there should only be 0 numbers of terms on the right
    # Example: f = 3n + 1  10n (2 terms on left, only one on the right)


# To prove f(n) = O(n) (that f(n) is asymptotically  ())
    # Video reference: https://www.youtube.com/watch?v=A03oI0znAoc
    # Note reference: http://www.columbia.edu/~cs2035/courses/csor4231.F11/mathbasics.pdf

# Example 1: 
# f = 2n + 3  10n (10 being c and n being ())

# Create a figure.
fig, ax = plt.subplots(figsize=(12, 6))

# n values.
n = np.linspace(0.0, 3.0, 1000)

# Plot f(n).
ax.plot(n, (2*n)+3, label="$2n + 3$")

# Plot cg(n).
ax.plot(n, 10*n, label="$10n$")

# Legend.
ax.legend()

# Show.
plt.show();
```

<h2 style="color: rgb(0, 91, 94);">Turing Machine</h2>

<hr style="border-top: 1px solid rgb(0, 91, 94);" />

```python
# State table.
table = {
    ('X', '_'): ['_', 'R', 'T'],
    ('X', '0'): ['0', 'R', 'X'],
    ('X', '1'): ['1', 'R', 'Y'],
    ('Y', '_'): ['_', 'R', 'F'],
    ('Y', '0'): ['0', 'R', 'Y'],
    ('Y', '1'): ['1', 'R', 'X'],
}

# Tape input.
tape = list('01110001')
# Position on tape.
pos = 0
# Initial state is first in table.
state = 'X'

# Keep going while we are not in a halting state.
while state not in ['T', 'F']:
    # Print the current status.
    print(''.join(tape[:pos]) + state + ''.join(tape[pos:]))
    # Get the row of the table.
    row = table[(state, tape[pos])]
    # Overwrite the symbol.
    tape[pos] = row[0]
    # Move left or right.
    if row[1] == 'R':
        # Put blanks on tape as necessary.
        if pos == len(tape) - 1:
            tape = tape + ['_']
        # Increase position.
        pos = pos + 1
    else:
        # Put blanks on tape as necessary.
        if pos == 0:
            tape = ['_'] + tape
            # The position on the tape has to move with it.
            pos = pos + 1
        # Increase position.
        pos = pos - 1
    # Update the state.
    state = row[2]

# Print the current status.
print(''.join(tape[:pos]) + state + ''.join(tape[pos:]))
```
