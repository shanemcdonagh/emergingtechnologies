```python
import qiskit
```

```python
circuit = qiskit.QuantumCircuit(4,1)
```

```python
circuit.draw(output="mpl")
```

```python
circuit.x(0)
circuit.x(1)
```

```python
circuit.draw(output="mpl")
```

```python
circuit.draw(output="mpl")
```

```python
circuit.barrier()
```

```python
circuit.draw(output="mpl")
```

```python
circuit.cnot(0,2)
```

```python
circuit.draw(output="mpl")
```

```python
circuit.cx(1,2)
```

```python
circuit.draw(output="mpl")
```

```python
circuit.ccx(0,1,3)
```

```python
circuit.draw(output="mpl")
```

```python
circuit.barrier()
```

```python
circuit.draw(output="mpl")
```

```python
circuit.measure(2,0)
```

```python
circuit.draw(output="mpl")
```

```python
circuit.measure(3,0)
```

```python
circuit.draw(output="mpl")
```

```python
sim = qiskit.Aer.get_backend('aer_simulator')
result = sim.run(circuit).result()
counts = result.get_counts()
qiskit.visualization.plot_histogram(counts)
```
